```python
import os
os.environ['TF_CPP_MIN_LOG_LEVEL'] = '2'

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.utils import to_categorical
from sklearn.model_selection import StratifiedKFold
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt

np.random.seed(777)
tf.random.set_seed(777)

data = np.loadtxt('glass.csv', delimiter=',')

x_data = data[:, 0:10]
y_data = data[:, 10]-1
y_encoded = to_categorical(y_data)

n_fold = 5
skf = StratifiedKFold(n_splits=n_fold, shuffle=True, random_state=48)

accuracy = []
for train, test in skf.split(x_data,y_data):
    model = Sequential()
    model.add(Dense(20, input_dim=10, activation='relu'))
    model.add(Dense(15, activation='relu'))
    model.add(Dense(7, activation='softmax'))
    model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
    model.summary()
    y_encoded_train = to_categorical(y_data[train])
    y_encoded_test = to_categorical(y_data[test])
    model.fit(x_data[train], y_encoded_train, epochs=50, batch_size=10)
    train_result = model.evaluate(x_data[train], y_encoded_train)
    result = model.evaluate(x_data[test], y_encoded_test)
    print("\n Test loss: %.4f, Accuracy: %.4f" % (result[0], result[1]))
    k_accuracy = "%.3f" % (result[1])
    accuracy.append(k_accuracy)
print("\n %.f fold accuracy:" % n_fold, accuracy)
```
